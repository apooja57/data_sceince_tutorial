# Analyze `sample1`

Load `all_samples_data.csv`, select only `sample1`, and calculate the statistics needed for a z-test.

```python
import pandas as pd

data = pd.read_csv("new_all_samples_data.csv")
sample1 = data["sample1"].dropna()

sample1.head()
```

```python
sample1_stats = pd.Series({
    "n": sample1.count(),
    "mean": sample1.mean(),
    "variance": sample1.var(ddof=1),
    "std_dev": sample1.std(ddof=1),
    "standard_error": sample1.std(ddof=1) / sample1.count() ** 0.5,
})

sample1_stats
```

```python
hypothesized_mean = 80  # Change this value for your z-test null hypothesis.

z_score = (sample1_stats["mean"] - hypothesized_mean) / sample1_stats["standard_error"]
z_score


```

```python
import pandas as pd

df = pd.read_csv("all_samples_data.csv")
df.head()
```

```python
means = df.mean()
variances = df.var()  # sample variance, uses n - 1

result = pd.DataFrame({
    "mean": means,
    "variance": variances,
})
result
```

```python
overall_mean = df.to_numpy().mean()
overall_variance = df.stack().var()

print(f"Overall mean: {overall_mean:.6f}")
print(f"Overall sample variance: {overall_variance:.6f}")
```
